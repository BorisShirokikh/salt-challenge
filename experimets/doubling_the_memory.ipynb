```python
import numpy as np
import pandas as pd
```

```python
from saltsegm.torch_models.unet import UNet
```

```python
model = UNet(n_channels=1, n_classes=1,
             n_filters=64, dropout_rate=0.1)
```

```python
model.cuda()
```

```python
import torch
import torch.nn as nn

from torch.nn import BCEWithLogitsLoss
from torch.optim import Adam

loss_fn = BCEWithLogitsLoss()
optimizer = Adam(model.parameters())
```

```python
x = np.random.rand(32, 1, 128, 128)
x = np.array(x, dtype='float32')
x = torch.from_numpy(x).cuda()

y = np.random.rand(32, 1, 128, 128)
y = np.array(y, dtype='float32')
y = torch.from_numpy(y).cuda()
```

```python
model.train()
```

```python
pred = model(x)
loss = loss_fn(pred, y)

optimizer.zero_grad()
loss.backward()
optimizer.step()
```
